```python
prefix= 'C:/Users/poona/Downloads/data_Q1_2019/drive_stats_2019_Q1/'
```

```python
paths = [prefix + '2019-01-01.csv',
prefix + '2019-01-02.csv',
prefix + '2019-01-03.csv',
prefix + '2019-01-04.csv',
prefix + '2019-01-05.csv',
prefix + '2019-01-06.csv',
prefix + '2019-01-07.csv',
prefix + '2019-01-08.csv',
prefix + '2019-01-09.csv',
prefix + '2019-01-10.csv']

data = spark.read.csv(paths, header='true')
data.createOrReplaceTempView("data")

query = "SELECT DISTINCT date, serial_number FROM data"
data2 = spark.sql(query)
data2.createOrReplaceTempView("data2")

reservoir=[]
K=50000
counter = 0

split_weights = [1.0] * 1 #Just making it iterable
splits = data2.randomSplit(split_weights)

for i in range(len(splits)):
    temp=splits[i].collect()
    for j in temp:
        if counter<50000:
            reservoir.append(j)
            counter = counter + 1
        else:
            prob = 50000/counter
            counter = counter + 1
            if random.random() < prob:
                reservoir[random.choice(range(50000))] = j
```
